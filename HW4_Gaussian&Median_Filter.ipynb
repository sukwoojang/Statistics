```python
# Image
from scipy import misc
face = misc.face(gray=True)


face.shape
```

```python
import matplotlib.pyplot as plt

plt.imshow(face, cmap = plt.cm.gray)
```

```python
import numpy as np

noisy_face = np.copy(face).astype(np.float)
noisy_face += face.std() * 0.5 * np.random.standard_normal(face.shape)

plt.imshow(noisy_face, plt.cm.gray)
```

```python
# gaussian
    
def gaussian(image):
    new_img = np.copy(image)
    gaussian_filter = np.array([[1,2,1],
                      [2,4,2],
                      [1,2,1]]) / 16
    print(gaussian_filter)
    x = image.shape[0]
    y = image.shape[1]
    for i in range(1,x-1):
        for j in range(1,y-1):
            s = sum(new_img[i-1:i+2,j-1:j+2])
            new_img[i-1:i+2,j-1:j+2] = s * new_img[i-1:i+2,j-1:j+2] * gaussian_filter
    return new_img
    
gaussian_face = gaussian(noisy_face)
plt.imshow(gaussian_face,)
```

```python
# Median

def median(image):
    x = image.shape[0]
    y = image.shape[1]
    
    new_img = np.copy(image)
    
    for i in range(1, x-1):
        for j in range(1, y-1):
            medianvalue = np.median(image[i-1:i+2,j-1:j+2])
            new_img[i,j] = medianvalue
            
    return new_img


median_face = median(noisy_face)
plt.imshow(median_face, plt.cm.gray)
```

```python
from scipy import ndimage
shifted_face = ndimage.shift(face, (50, 50))
shifted_face2 = ndimage.shift(face, (50, 50), mode='nearest')
rotated_face = ndimage.rotate(face, 30)
cropped_face = face[50:-50, 50:-50]
zoomed_face = ndimage.zoom(face, 4)
```
